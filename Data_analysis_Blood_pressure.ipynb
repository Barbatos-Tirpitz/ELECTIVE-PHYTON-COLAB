# 66-80. Data Analysis with CSV Files:
Load a CSV file, calculate basic statistics like mean and
standard deviation, and create visualizations with matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "/content/blood_pressure.csv"
data = pd.read_csv(file_path)

# Display basic informatiod
print(data.info())

# mean and standard deviation
mean_before = data['bp_before'].mean()
std_before = data['bp_before'].std()

mean_after = data['bp_after'].mean()
std_after = data['bp_after'].std()

print(f'Mean BP Before: {mean_before}, Std Dev BP Before: {std_before}')
print(f'Mean BP After: {mean_after}, Std Dev BP After: {std_after}')

# visualizations
# Line plot
plt.figure(figsize=(10, 6))
plt.plot(data['patient'], data['bp_before'], label='Before', marker='o')
plt.plot(data['patient'], data['bp_after'], label='After', marker='o')
plt.title('Blood Pressure Before and After')
plt.xlabel('Patient')
plt.ylabel('Blood Pressure')
plt.legend()
plt.show()

#  Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(data['bp_before'], data['bp_after'])
plt.title('Scatter Plot of Blood Pressure Before and After')
plt.xlabel('Blood Pressure Before')
plt.ylabel('Blood Pressure After')
plt.show()
```
